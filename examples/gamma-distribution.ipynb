# Gamma distribution sandbox
This notebook helps to explore the Gamma distribution, its features, and using it as a probability distribution generating values from 0 to 1, and then using it for weighted distribution around a desired mean. This will be useful to infectious disease simulations like OpenABM-Covid19.
## Gamma distribution
First let's look at the default gamma distribution available in scipy.stats.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma 

# plt.plot(values, label='Scipy Default', color='green')

# #add legend
# plt.legend(title='Settings')

# #add axes labels and a title
# plt.ylabel('Frequency', fontsize=14)
# plt.xlabel('Probability', fontsize=14)
# plt.title('Gamma Distribution', fontsize=16)

# plt.show()

a = 1 # shape=1 (1,1 is basically the inverse exponential distribution)

fig, ax = plt.subplots(1, 1)
x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a), 'r-', lw=3, alpha=0.6, label='gamma pdf (shape=1)')

# r = gamma.rvs(a, size=1000)
# ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)

# Now add shape=2
a2 = 2
x2 = np.linspace(gamma.ppf(0.01, a2), gamma.ppf(0.99, a2), 100)
ax.plot(x2, gamma.pdf(x2, a2), 'b-', lw=3, alpha=0.6, label='gamma pdf (shape=2)')

# Now add shape=3
a3 = 3
x3 = np.linspace(gamma.ppf(0.01, a3), gamma.ppf(0.99, a3), 100)
ax.plot(x3, gamma.pdf(x3, a3), 'g-', lw=3, alpha=0.6, label='gamma pdf (shape=3)')

# Now add shape=2
a4 = 2
r4 = 1 / 6
x4 = np.linspace(gamma.ppf(0.01, a4, scale = r4), gamma.ppf(0.99, a4, scale = r4), 100)
ax.plot(x4, gamma.pdf(x4, a4, scale = r4), 'y-', lw=3, alpha=0.6, label='gamma pdf (shape=2, scale=1/6)')

ax.legend(loc='best', frameon=False)

plt.show()
```

## Change loc
Change the loc value and see where the chart moves to

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma 

a = 1 # shape=1 (1,1 is basically the inverse exponential distribution)

fig, ax = plt.subplots(1, 1)

# Now add shape=2, loc = 0
a = 2
x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a), 'b-', lw=3, alpha=0.6, label='gamma pdf (shape=2, loc=0)')

# Now add shape=2, loc = -0.5
a2 = 2
x2 = np.linspace(gamma.ppf(0.01, a2), gamma.ppf(0.99, a2), 100)
ax.plot(x2, gamma.pdf(x2, a2, loc = -0.5), 'g-', lw=3, alpha=0.6, label='gamma pdf (shape=2, loc=-0.5)')


# Now add shape=2, loc = 0, scale = 0.5
a3 = 2
scale3 = 0.5
x3 = np.linspace(gamma.ppf(0.01, a3, scale = scale3), gamma.ppf(0.99, a3, scale = scale3), 100)
ax.plot(x3, gamma.pdf(x3, a3, scale = scale3), 'y-', lw=3, alpha=0.6, label='gamma pdf (shape=2, scale = 0.5)')

ax.legend(loc='best', frameon=False)

plt.show()

mean = gamma.mean(a)
mean2 = gamma.mean(a2)
mean3 = gamma.mean(a3)
print("Mean 1:",mean)
print("Mean 2:",mean2)
print("Mean 3:",mean3)
```
